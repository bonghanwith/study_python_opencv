### ORB (Oriented FAST and Rotated BRIEF)
SIFT and SURF are good in what they do, but what if you have to pay a few dollars every year to use them in your applications? Yeah, they are patented!!! To solve that problem, OpenCV devs came up with a new "FREE" alternative to SIFT & SURF, and that is ORB.

#### Theory
OpenCV 애호가로서 ORB에서 가장 중요한 점은 "OpenCV Labs"에서 나온 것입니다. 이 알고리즘은 Ethan Rublee, Vincent Rabaud, Kurt Konolige 및 Gary R. Bradski가 자신의 논문 ORB 에서 가져 왔습니다 . 2011 년 SIFT 또는 SURF 에 대한 효율적인 대안 입니다. 제목에서 알 수 있듯이 SIFT 및 SURF 계산의 좋은 대안입니다 비용, 매칭 성능 및 주로 특허. 그렇습니다. SIFT와 SURF는 특허를 받았으며 사용을 위해 지불해야합니다. 하지만 ORB는 그렇지 않습니다 !!!

ORB는 기본적으로 FAST 키포인트 검출기와 성능 향상을위한 다양한 수정 사항이 포함 된 BRIEF 설명 자의 융합입니다. 먼저 FAST를 사용하여 키포인트를 찾은 다음 해리스 코너 측정을 적용하여 그 중 상위 N 포인트를 찾습니다. 또한 피라미드를 사용하여 멀티 스케일 기능을 생성합니다. 그러나 한 가지 문제는 FAST가 방향을 계산하지 않는다는 것입니다. 회전 불변성은 어떨까요? 저자는 다음과 같이 수정했다.

중앙에 위치한 구석이있는 패치의 강도 가중치 중심을 계산합니다. 이 구석 점에서 중심까지의 벡터의 방향은 방향을 나타냅니다. 회전 불변량을 향상시키기 위해 모멘트는 반경 r의 원형 영역에 있어야하는 x와 y로 계산됩니다. 여기서 r은 패치의 크기입니다.

이제 기술자의 경우 ORB는 간단한 설명자를 사용합니다. 그러나 우리는 이미 BRIEF가 회전과 함께 제대로 작동하지 않는 것을 보았습니다. ORB가하는 일은 키포인트의 방향에 따라 BRIEF를 "조종"하는 것입니다. 위치 \\((x_i, y_i)\\)에있는 n 개의 이진 테스트의 모든 기능 세트에 대해이 픽셀의 좌표를 포함하는 \\(2 \times n\\) 매트릭스 \\(S\\)를 정의하십시오. 그런 다음 패치의 방향 \\(θ\\)를 사용하여 회전 행렬을 찾고 \\(S\\)를 회전시켜 조향 된 (회전 된) 버전 \\(S_θ\\)를 얻습니다.

ORB는 각도를 \\(2 \pi /30\\)(12도) 증가분으로 이산화하고 미리 계산 된 BRIEF 패턴의 찾아보기 테이블을 만듭니다. 키포인트 방향 \\(θ\\)가 뷰 전반에 걸쳐 일치하는 한 올바른 점 \\(S_θ\\) 세트가 해당 설명자를 계산하는 데 사용됩니다.

브리핑은 각 비트 피처가 큰 분산을 가지며 평균이 0.5에 가까운 중요한 속성을 갖고 있습니다. 그러나 일단 키포인트 방향을 따라 지향되면이 속성을 잃고 분산됩니다. 높은 분산은 입력에 차 동적으로 응답하기 때문에 기능을 더욱 차별화합니다. 또 하나의 바람직한 특성은 각 테스트가 그 결과에 기여할 것이기 때문에 테스트를 상관시키지 않는 것입니다. 이 모든 문제를 해결하기 위해 ORB는 모든 가능한 바이너리 테스트 중에서 욕심이 많은 검색을 실행하여 분산이 크고 평균 0.5에 가까울뿐만 아니라 상관 관계가없는 것을 찾습니다. 그 결과를 **rBRIEF** 라고 합니다 .

디스크립터 매칭을 위해 전통적인 LSH를 개선 한 멀티 프로브 LSH가 사용됩니다. 이 신문에 따르면 ORB는 SURF보다 훨씬 빠르며 SIFT와 ORB 디스크립터는 SURF보다 잘 작동한다고합니다. ORB는 파노라마 스티칭 (panorama stitching) 등을위한 저전력 디바이스에서 좋은 선택입니다.

#### ORB in OpenCV
통상, 우리는 [cv.ORB()](https://docs.opencv.org/3.4/db/d95/classcv_1_1ORB.html) 함수를 사용하거나 feature2d 공통 인터페이스 를 사용하여 ORB 객체를 생성해야합니다 . 여러 가지 선택적 매개 변수가 있습니다. 가장 유용한 것들은 보유 할 피쳐의 최대 수 (디폴트는 500)를 나타내는 nFeatures, 해리스 스코어 또는 피스의 순위를 매기기위한 FAST 스코어 (기본적으로 해리스 스코어) 등을 나타내는 scoreType입니다. 또 다른 매개 변수 인 WTA_K는 포인트 수를 결정합니다 방향성을 가지는 BRIEF descriptor의 각 요소를 생성합니다. 기본적으로 두 개입니다. 즉 한 번에 두 개의 점을 선택합니다. 이 경우 일치시키기 위해 NORM_HAMMING 거리가 사용됩니다. WTA_K가 3 또는 4 인 경우 간단한 디스크립터를 생성하기 위해 3 또는 4 포인트가 소요되며 일치 거리는 NORM_HAMMING2에 의해 정의됩니다.

Below is a simple code which shows the use of ORB.

```python
%matplotlib inline
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

img = cv.imread('simple.jpg', 0)

# Initiate ORB detector
orb = cv.ORB_create()

# find the keypoints with ORB
kp = orb.detect(img, None)

# compute the descriptors with ORB
kp, des = orb.compute(img, kp)

# draw only keypoints location,not size and orientation
img2 = cv.drawKeypoints(img, kp, None, color = (0, 255, 0), flags = 0)
plt.imshow(img2), plt.show()
```
