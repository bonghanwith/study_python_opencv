#### Theory
In last chapter, we saw Harris Corner Detector. Later in 1994, J. Shi and C. Tomasi made a small modification to it in their paper Good Features to Track which shows better results compared to Harris Corner Detector. The scoring function in Harris Corner Detector was given by:  

\\( R = \lambda_1 \lambda_2 - k(\lambda_1+\lambda_2)^2 \\)  
Instead of this, Shi-Tomasi proposed:

\\( R = min(λ_1, λ_2) \\)  
If it is a greater than a threshold value, it is considered as a corner. If we plot it in \\(λ_1 − λ_2\\) space as we did in Harris Corner Detector, we get an image as below:
![](shitomasi_space.png)
From the figure, you can see that only when \\(λ_1\\) and \\(λ_2\\) are above a minimum value, \\(λ_{min}\\), it is conidered as a corner(green region).

#### Code
OpenCV에는 cv.goodFeaturesToTrack () 함수가 있습니다. Shi-Tomasi 방법 (또는 지정한 경우 Harris Corner Detection)을 사용하여 이미지에서 N 개의 가장 강한 모서리를 찾습니다. 일반적으로 이미지는 회색 음영 이미지 여야합니다. 그런 다음 원하는 모서리 수를 지정하십시오. 그런 다음 품질 수준을 지정합니다. 품질 수준은 0-1 사이의 값으로,이 값 아래에서 모든 사람이 거부되는 모퉁이의 최소 품질을 나타냅니다. 그런 다음 감지 된 구석 사이의 최소 유클리드 거리를 제공합니다.

이 모든 정보를 통해이 함수는 이미지의 모퉁이를 찾습니다. 품질 수준 이하의 모든 구석이 거부됩니다. 그런 다음 품질에 따라 내림차순으로 나머지 모서리를 정렬합니다. 그런 다음 함수는 가장 강한 모서리를 가져 와서 최소 거리의 범위에서 가까운 모든 모서리를 버리고 N 가장 강한 모서리를 반환합니다.

아래 예에서 우리는 25 개의 최고의 모서리를 찾으려고 노력할 것입니다.

```python
%matplotlib inline
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

img  = cv.imread('blox.jpg')
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

corners = cv.goodFeaturesToTrack(gray, 25, 0.01, 10)
corners = np.int0(corners)

for i  in corners:
    x, y = i.ravel()
    cv.circle(img, (x, y), 3, 255, -1)
    
plt.imshow(img)
plt.show()
```
