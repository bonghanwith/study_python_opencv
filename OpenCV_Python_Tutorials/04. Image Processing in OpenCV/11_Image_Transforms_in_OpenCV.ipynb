##### 이론
푸리에 변환은 다양한 필터의 주파수 특성을 분석하는 데 사용됩니다. 이미지의 경우 2D 이산 푸리에 변환 (DFT)을 사용하여 주파수 도메인을 찾습니다. FFT (Fast Fourier Transform)라고하는 빠른 알고리즘이 DFT 계산에 사용됩니다. 이들에 대한 세부 사항은 모든 이미지 처리 또는 신호 처리 교재에서 찾을 수 있습니다. 추가 리소스 _ 섹션을 참조하십시오.

사인파 신호의 경우, x (t) = Asin (2πft)에 대해, f는 신호의 주파수라고 말할 수 있으며, 주파수 도메인을 취하면 f에서의 스파이크를 볼 수 있습니다. 신호가 이산 신호로 샘플링되면 동일한 주파수 도메인을 갖지만 [-π, π] 또는 [0,2π] (N 포인트 DFT의 경우 [0, N]) 범위에서 주기적입니다. 이미지를 두 방향으로 샘플링 된 신호로 간주 할 수 있습니다. 따라서 X와 Y 방향 모두에서 푸리에 변환을하면 이미지의 주파수 표현을 얻을 수 있습니다.

보다 직관적으로 정현파 신호의 경우 진폭이 짧은 시간에 매우 빠르게 변하는 경우 고주파 신호라고 말할 수 있습니다. 천천히 변화하는 경우 저주파수 신호입니다. 동일한 아이디어를 이미지로 확장 할 수 있습니다. 이미지에서 진폭은 크게 다릅니다. 가장자리 지점 또는 소음. 그래서 우리는 가장자리와 노이즈가 이미지의 고주파수 내용이라고 말할 수 있습니다. 진폭에 큰 변화가 없다면 저주파 성분입니다. (일부 링크는 추가 리소스 _에 추가됩니다. 예를 통해 직관적으로 주파수 변환을 설명합니다).

이제 푸리에 변환을 찾는 방법을 살펴 보겠습니다.

##### Numpy의 푸리에 변환

먼저 Numpy를 사용하여 Fourier Transform을 찾는 방법을 살펴 보겠습니다. Numpy는 이것을 수행하기위한 FFT 패키지를 가지고 있습니다. np.fft.fft2 () 는 복잡한 배열이 될 주파수 변환을 제공합니다. 첫 번째 인수는 그레이 스케일 인 입력 이미지입니다. 두 번째 인수는 출력 배열의 크기를 결정하는 옵션입니다. 입력 영상의 크기보다 크다면 FFT 계산 전에 입력 영상에 0을 덧붙입니다. 입력 이미지보다 작 으면 입력 이미지가 잘립니다. 인수가 전달되지 않으면 출력 배열 크기는 입력과 같습니다.

이제 결과를 얻었 으면 제로 주파수 구성 요소 (DC 구성 요소)가 왼쪽 상단에 표시됩니다. 중심으로 가져 오려면 양쪽 방향으로 N / 2 만큼 결과를 이동해야합니다 . 이는 단순히 np.fft.fftshift () 함수로 수행됩니다 . (분석하기가 더 쉽습니다.) 주파수 변환을 찾으면 크기 스펙트럼을 찾을 수 있습니다. 
.

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

img = cv.imread('messi5.jpg',0)

f      = np.fft.fft2(img)
fshift = np.fft.fftshift(f)

magnitude_spectrum = 20 * np.log(np.abs(fshift))

plt.subplot(121),plt.imshow(img, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(magnitude_spectrum, cmap = 'gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])

plt.show()
```

보시다시피, 중심에서 더 낮은 백분율의 컨텐츠를 보여주는 더 하얀 영역을 볼 수 있습니다.

그래서 당신은 주파수 변환을 발견했습니다. 이제 하이 패스 필터링과 같은 주파수 영역에서 몇 가지 작업을 수행하고 이미지를 재구성합니다. 즉 역 DFT를 찾습니다. 이를 위해서는 크기가 60x60 인 직사각형 윈도우로 마스킹하여 저주파를 제거하면됩니다. 그런 다음 np.fft.ifftshift ()를 사용하여 역 이동을 적용하여 DC 구성 요소가 다시 왼쪽 상단 모서리에 오게하십시오. 그런 다음 np.ifft2 () 함수를 사용하여 역 FFT를 찾으십시오 . 결과는 다시 복소수가됩니다. 절대 가치를 취할 수 있습니다.

See, You can see more whiter region at the center showing low frequency content is more.

So you found the frequency transform Now you can do some operations in frequency domain, like high pass filtering and reconstruct the image, ie find inverse DFT. For that you simply remove the low frequencies by masking with a rectangular window of size 60x60. Then apply the inverse shift using np.fft.ifftshift() so that DC component again come at the top-left corner. Then find inverse FFT using np.ifft2() function. The result, again, will be a complex number. You can take its absolute value.

```python
rows, cols = img.shape
crow, ccol = int(rows / 2), int(cols / 2)

fshift[crow - 30:crow + 30, ccol - 30:ccol + 30] = 0
f_ishift = np.fft.ifftshift(fshift)
img_back = np.fft.ifft2(f_ishift)
img_back = np.abs(img_back)

plt.subplot(131),plt.imshow(img, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(img_back, cmap = 'gray')
plt.title('Image after HPF'), plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.imshow(img_back)
plt.title('Result in JET'), plt.xticks([]), plt.yticks([])
plt.show()
```

결과는 하이 패스 필터링이 에지 감지 작업임을 나타냅니다. 이것이 이미지 그라디언트 장에서 본 것입니다. 이것은 또한 대부분의 이미지 데이터가 스펙트럼의 저주파수 영역에 있음을 보여줍니다. 어쨌든 우리는 Numpy에서 DFT, IDFT 등을 찾는 법을 보았습니다. 이제 OpenCV에서 수행하는 방법을 살펴 보겠습니다.

결과, 특히 JET 색상의 마지막 이미지를 자세히 보면, 일부 유물 (빨간색 화살표로 표시된 인스턴스 하나)을 볼 수 있습니다. 그곳에는 리플과 같은 구조가 나타나는데, 링잉 효과 라고 합니다 . 이것은 마스킹에 사용 된 직사각형 창 때문에 발생합니다. 이 마스크는이 문제를 일으키는 sinc 모양으로 변환됩니다. 따라서 직사각형 창은 필터링에 사용되지 않습니다. 더 나은 옵션은 Gaussian Windows입니다.

he result shows High Pass Filtering is an edge detection operation. This is what we have seen in Image Gradients chapter. This also shows that most of the image data is present in the Low frequency region of the spectrum. Anyway we have seen how to find DFT, IDFT etc in Numpy. Now let's see how to do it in OpenCV.

If you closely watch the result, especially the last image in JET color, you can see some artifacts (One instance I have marked in red arrow). It shows some ripple like structures there, and it is called ringing effects. It is caused by the rectangular window we used for masking. This mask is converted to sinc shape which causes this problem. So rectangular windows is not used for filtering. Better option is Gaussian Windows.

##### OpenCV의 푸리에 변환
OpenCV는이를 위해 cv.dft () 및 cv.idft () 함수를 제공합니다 . 이전과 동일한 결과를 반환하지만 두 개의 채널을 반환합니다. 첫 번째 채널은 결과의 실수 부분을 가지며 두 번째 채널은 결과의 허수 부분을 갖습니다. 입력 이미지는 먼저 np.float32로 변환되어야합니다. 우리는 그것을하는 방법을 보게 될 것입니다.

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

%matplotlib inline

img = cv.imread('messi5.jpg',0)
dft = cv.dft(np.float32(img),flags = cv.DFT_COMPLEX_OUTPUT)
dft_shift = np.fft.fftshift(dft)
magnitude_spectrum = 20 * np.log(cv.magnitude(dft_shift[:, :, 0], dft_shift[:, :, 1]))

plt.subplot(121),plt.imshow(img, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(magnitude_spectrum, cmap = 'gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()
```

> **노트** 단일 샷에서 크기와 위상을 모두 반환 하는 cv.cartToPolar () 를 사용할 수도 있습니다

이제 역 DFT를 수행해야합니다. 이전 세션에서는 HPF를 만들었습니다. 이번에는 이미지에서 고주파수 내용을 제거하는 방법, 즉 LPF를 이미지에 적용하는 방법을 살펴 보겠습니다. 그것은 실제로 이미지를 흐리게 만듭니다. 이를 위해 저주파수에서 높은 값 (1)을 갖는 마스크를 먼저 만듭니다. 즉, HF 영역에서 LF 콘텐츠를 전달하고 HF 영역에서는 0을 전달합니다.

```python
rows, cols = img.shape
crow, ccol = int(rows / 2) , int(cols / 2)

# create a mask first, center square is 1, remaining all zeros
mask = np.zeros((rows, cols,2), np.uint8)
mask[crow - 30:crow + 30, ccol - 30:ccol + 30] = 1

# apply mask and inverse DFT
fshift   = dft_shift * mask
f_ishift = np.fft.ifftshift(fshift)
img_back = cv.idft(f_ishift)
img_back = cv.magnitude(img_back[:,:,0],img_back[:,:,1])

plt.subplot(121),plt.imshow(img, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(img_back, cmap = 'gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()
```

> 노트  
보통 OpenCV 함수 cv.dft () 와 cv.idft () 는 Numpy보다 빠릅니다. 그러나 Numpy 기능은 사용자에게 더욱 친숙합니다. 성능 문제에 대한 자세한 내용은 아래 섹션을 참조하십시오.

##### DTF의 성능 최적화
DFT 계산의 성능은 일부 배열 크기에서 더 좋습니다. 배열 크기가 2의 배수 일 때 가장 빠릅니다. 크기가 2, 3 및 5의 곱인 배열도 매우 효율적으로 처리됩니다. 따라서 코드의 성능이 걱정되면 DFT를 찾기 전에 배열의 크기를 최적의 크기 (0으로 패딩)로 수정할 수 있습니다. OpenCV의 경우 수동으로 0을 덧붙여 야합니다. 그러나 Numpy의 경우 새로운 크기의 FFT 계산을 지정하면 자동으로 0이 추가됩니다.

그렇다면이 최적의 크기를 어떻게 찾습니까? OpenCV는 이것을 위해 cv.getOptimalDFTSize () 함수를 제공합니다 . cv.dft () 와 np.fft.fft2 () 모두에 적용됩니다 . IPython magic command timeit을 사용하여 성능을 확인해 봅시다.

```python
rows,cols = img.shape
print("원본 사이즈   : {} {}".format(rows,cols))

nrows = cv.getOptimalDFTSize(rows)
ncols = cv.getOptimalDFTSize(cols)
print("변환된 사이즈 : {} {}".format(nrows,ncols))
```

크기 (342,548)는 (360, 576)으로 수정됩니다. 이제는 OpenCV 용으로 0으로 채우고 DFT 계산 성능을 찾으십시오. 새로운 큰 제로 배열을 만들고 데이터를 복사하거나 cv.copyMakeBorder ()를 사용 하면 됩니다.

```python
nimg = np.zeros ((nrows, ncols))
nimg [: rows, : cols] = img

# or
'''
right      = ncols - cols
bottom     = nrows - rows
bordertype = cv.BORDER_CONSTANT #just to avoid line breakup in PDF file
nimg       = cv.copyMakeBorder(img, 0, bottom, 0, right,bordertype, value = 0)
'''
```

이제 Numpy 함수의 DFT 성능 비교를 계산합니다.

```python
%timeit fft1 = np.fft.fft2(img)
%timeit fft2 = np.fft.fft2(img,[nrows,ncols])
```

4x 속도 향상을 보여줍니다. 이제 우리는 OpenCV 기능을 사용하여 똑같이 시도 할 것입니다.

```python
% timeit dft1 = cv.dft (np.float32 (img), flags = cv.DFT_COMPLEX_OUTPUT)
% timeit dft2 = cv.dft (np.float32 (nimg), flags = cv.DFT_COMPLEX_OUTPUT)
```

그것은 또한 4x 속도 향상을 보여줍니다. 또한 OpenCV 기능이 Numpy 기능보다 약 3 배 빠르다는 것을 알 수 있습니다. 이것은 역 FFT에 대해서도 테스트 할 수 있으며, 이것은 당신을위한 연습으로 남겨 두었습니다.

##### Laplacian이 하이 패스 필터 인 이유는 무엇입니까?

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

# simple averaging filter without scaling parameter
mean_filter = np.ones((3,3))

# creating a gaussian filter
x = cv.getGaussianKernel(5,10)
gaussian = x * x.T

# different edge detecting filters
# scharr in x-direction
scharr = np.array([[-3, 0, 3],
                   [-10,0,10],
                   [-3, 0, 3]])

# sobel in x direction
sobel_x= np.array([[-1, 0, 1],
                   [-2, 0, 2],
                   [-1, 0, 1]])

# sobel in y direction
sobel_y= np.array([[-1,-2,-1],
                   [0, 0, 0],
                   [1, 2, 1]])

# laplacian
laplacian=np.array([[0, 1, 0],
                    [1,-4, 1],
                    [0, 1, 0]])

filters = [mean_filter, gaussian, laplacian, sobel_x, sobel_y, scharr]
filter_name = ['mean_filter', 'gaussian','laplacian', 'sobel_x', \
                'sobel_y', 'scharr_x']


'''
f      = np.fft.fft2(img)
fshift = np.fft.fftshift(f)
magnitude_spectrum = 20 * np.log(np.abs(fshift))


dft = cv.dft(np.float32(img),flags = cv.DFT_COMPLEX_OUTPUT)
dft_shift = np.fft.fftshift(dft)
magnitude_spectrum = 20 * np.log(cv.magnitude(dft_shift[:, :, 0], dft_shift[:, :, 1]))

fft_filters = [ cv.dft(np.float32(x),flags = cv.DFT_COMPLEX_OUTPUT) for x in filters]
fft_shift = [np.fft.fftshift(y) for y in fft_filters]
mag_spectrum = [20 * np.log(cv.magnitude(z[:, :, 0], z[:, :, 1])) for z in fft_shift]
'''

fft_filters = [np.fft.fft2(x) for x in filters]
fft_shift = [np.fft.fftshift(y) for y in fft_filters]
mag_spectrum = [np.log(np.abs(z) + 1) for z in fft_shift]



for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(mag_spectrum[i],cmap = 'gray')
    plt.title(filter_name[i]), plt.xticks([]), plt.yticks([])
plt.show()

cv_fft_filters = [ cv.dft(np.float32(x),flags = cv.DFT_COMPLEX_OUTPUT) for x in filters]
cv_fft_shift = [np.fft.fftshift(y) for y in cv_fft_filters]
cv_mag_spectrum = [20 * np.log(cv.magnitude(z[:, :, 0], z[:, :, 1]) + 1) for z in cv_fft_shift]

for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(cv_mag_spectrum[i],cmap = 'gray')
    plt.title(filter_name[i]), plt.xticks([]), plt.yticks([])
plt.show()
```
